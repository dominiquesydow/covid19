# Molecule library

## Aim of this notebook

Quick start to explore molecule library.

```python
from pathlib import Path

import pandas as pd
from rdkit.Chem import PandasTools
```

```python
PROBIS_FOLDER = Path('.') / '..' / 'data' / 'probis' / 'probis_pocket_15_0.5'
MOLECULES_PATH = PROBIS_FOLDER / 'chembl_molecules_from_uniprot_ids.csv'
```

## Load molecule library

```python
dataset = pd.read_csv(MOLECULES_PATH)
dataset.shape
```

```python
dataset.head()
```

## Draw ligands

```python
PandasTools.AddMoleculeColumnToFrame(dataset, smilesCol='canonical_smiles')
```

```python
dataset[['ROMol', 'pIC50', 'molecule_chembl_id', 'target_chembl_id']].sort_values(
    'pIC50', 
    ascending=False
).head(20)
```
